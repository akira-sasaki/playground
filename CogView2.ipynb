# Run text2image

The file of text is specified by --input-source option: ***input.txt*** <br>
The folder of output image is specified by --output-path: ***samples_sat_v0.2***

Add ***--only-first-stage***, because GPU Memory size of Colab Pro+ is only 16GB.<br>
https://github.com/THUDM/CogView2/issues/8



````python
!python cogview2_text2image.py \
       --mode inference \
       --fp16 \
       --input-source input.txt \
       --output-path samples_sat_v0.2 \
       --batch-size 16 \
       --max-inference-batch-size 8 \
       --only-first-stage  \
       $@
````
